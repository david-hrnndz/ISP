```python
import numpy as np
import fractions
from Markov import Markov
```

```python
P = [[1/4, 3/4], [1, 0]]
lam = [1/3, 2/3]
x = Markov(lam, P)
```

```python
# Example 1: Calculate P(X_1 = 2)
x.prob(1, 2)
# Example 2: Calculate P(X_2 = 1, X_3 = 2)
# By conditional probability, this is equal to 
# P_1(X_3 = 2)*P(X_2 = 1)
# Where P_1(X_3 = 2) = P(X_3 = 2 | X_2 = 1)
x.prob_i(1, 3, 2)
fractions.Fraction(x.prob_i(1, 3, 2) * x.prob(2,1)).limit_denominator()
```

```python
## HW3_EX3
P_3_3 = [[0.2, 0.5, 0.3], [0.1, 0.1, 0.8], [0.5, 0.2, 0.3]]
lam_3_3 = [1/3, 1/3, 1/3]
X = Markov(lam_3_3, P_3_3)
print(f"P^3 = {np.linalg.matrix_power(P_3_3,3)}")
X.prob_i(2, 3, 2)
X.prob_i(1,3,3)
```
